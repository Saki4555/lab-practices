```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x(n):
    if 0 <= n < 10:
        return (n*n) + n
    elif -10 < n < 0:
        return (n*n) + n
    else:
        return 0
```

```python
def y(n):
    return 2 * x(-n - 5) * 2 * x(-n + 5)
```

```python
n_values = np.arange(-15, 16)
x = np.arange(-15,16
```

```python
x_values = [x(n) for n in n_values]
y_values = [y(n) for n in n_values]
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.stem(n_values, x_values, use_line_collection=True)
plt.title('$x(n)$')
plt.xlabel('$n$')
plt.ylabel('$x(n)$')

plt.subplot(2, 1, 2)
plt.stem(n_values, y_values, use_line_collection=True)
plt.title('$y(n)$')
plt.xlabel('$n$')
plt.ylabel('$y(n)$')

plt.tight_layout()
plt.show()
```
